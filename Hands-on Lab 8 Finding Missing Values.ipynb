The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
# 1. Dataset shape
print("Shape of the dataset (rows, columns):", df.shape)

# 2. Column info (data types, non-null counts)
print("\nDataset Info:")
print(df.info())

# 3. Summary statistics for numerical columns
print("\nSummary statistics (numerical columns):")
print(df.describe())

# 4. Summary statistics for categorical columns
print("\nSummary statistics (categorical columns):")
print(df.describe(include=['object']))
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
# Count of missing values in each column
missing_count = df.isnull().sum()

print(missing_count)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
# Set figure size
plt.figure(figsize=(15, 6))

# Heatmap of missing values
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")

plt.title("Heatmap of Missing Values", fontsize=16)
plt.show()                                                #✅ Yellow (or light color) = missing values
                                                          #✅ Purple/black (or dark color) = non-missing values
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
# Count missing rows in 'Employment' column
missing_employment = df['Employment'].isnull().sum()

print(f"Number of missing rows in Employment column: {missing_employment}")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
# Find the most frequent value in 'Employment' column
most_frequent_employment = df['Employment'].mode()[0]

print(f"The most frequent value in Employment column is: {most_frequent_employment}")

# Alternatively,,Shows frequency counts
employment_counts = df['Employment'].value_counts()

print("Employment value counts:\n", employment_counts)
print(f"\nMost frequent value: {employment_counts.idxmax()} ({employment_counts.max()} occurrences)")

```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
# Find the most frequent value in Employment
most_frequent_employment = df['Employment'].mode()[0]

# Fill missing values with the most frequent value
df['Employment'].fillna(most_frequent_employment, inplace=True)

# Verify that no missing values remain
print(df['Employment'].isnull().sum())
```
